### 눈금 맞춤 변경하기

matplotlib은 플롯에 등장하는 모든 것을 나타내는 파이썬 객체를 갖는 것을 목표로 한다.

예를 들어 firgure는 내부에 플롯 요소들이 등장하는 테두리 쳐진 박스라는 사실

각 matplotlib 객체는 하위 객체의 컨테이너 역할을 할 수도 있다.
- 예를 들어 각 figure는 하나 이상의 axes 객체를 포함할 수 있으며, 이 각각의 axes 객체는 다시 플롯 내용을 표시하는 다른 객체를 포함한다.

눈금 표시도 axes는 xaxis와 yaxis속성을 가지고 있으며 이 속성은 다시 축을 구성하는 선, 눈금, 레이블의 모든 특성을 포함하는 속성을 가진다.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('seaborn-whitegrid')

%matplotlib inline
```

### 주 눈금과 보조 눈금 

```python
ax = plt.axes(xscale='log', yscale='log')
```

```python
print(ax.xaxis.get_major_locator())
print(ax.xaxis.get_minor_locator())
```

```python
print(ax.xaxis.get_major_formatter())
print(ax.xaxis.get_minor_formatter())
```

```python
ax = plt.axes()
ax.plot(np.random.rand(50))
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

눈금이 없으면 유용한 경우?
- 여러 이미지를 그리드로 구성해 보여줄 때

```python
fig, ax = plt.subplots(5, 5, figsize=(5, 5))
fig.subplots_adjust(hspace=0, wspace=0)

# Get some face data from scikit-learn
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces().images

for i in range(5):
    for j in range(5):
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
        ax[i, j].imshow(faces[10 * i + j], cmap="bone")
```
