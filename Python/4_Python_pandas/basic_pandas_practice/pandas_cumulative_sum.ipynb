pandas cumulative sum
----
파이썬에서 판다스에서 cumulative sum 을 해보자
- 참고 문서 : https://stackoverflow.com/questions/22650833/pandas-groupby-cumulative-sum

```python
import pandas as pd
```

```python
file_path = 'Stock_Dataset(2017_07_06)/000020.csv'
df = pd.read_csv(file_path)
```

```python
df['Date'] = df['Date'].astype('datetime64[ns]')
```

```python
df.info()
```

```python
df.head()
```

```python
def get_year(value):
    year = value.year
    
    return year

def get_month(value):
    month = value.month
    
    return month
```

```python
df['year'] = df['Date'].apply(get_year)
df['month'] = df['Date'].apply(get_month)
```

```python
df.head()
```

```python
df.tail()
```

```python
print(df[df['year']==2005]['Open'].sum())
print(df[df['year']==2006]['Open'].sum())
print(df[df['year']==2007]['Open'].sum())
```

- 연도별 누적 합계를 구하기(항목 무관)
- 연도별 + 월별 누적 합계를 구하기(항목 무관)
- 항목은 편의상 Volumn(거래량)으로 하겠음

```python
# Company는 들어가지 않아도 되지만, 다른 항목이 있다고 가정하였음
# level=0으로 옵션을 줄 경우 단순히 연도별 합만 구해짐
# level=1을 줄 경우 연도별로 누적 합이 구해짐 -> 당연히 그렇게 됨...

df.groupby(['year', 'Company']).sum().groupby(level=0).cumsum().head()
```

```python
df.groupby(['year', 'Company']).sum().groupby(level=1).cumsum().head()
```

------

```python
df.groupby(['year', 'month', 'Company']).sum().head(10)
```

```python
# level=0으로 옵션을 주면 2005년을 기준으로 누적 합을, 2006년을 기준으로 누적 합을....이하 반복

df.groupby(['year', 'month', 'Company']).sum().groupby(level=0).cumsum()
```

```python
# 월별 누적 합계인데, 구하고자 하는 것은 아니다.

df.groupby(['year', 'month', 'Company']).sum().groupby(level=1).cumsum()
```

```python
def get_year_month(value):
    year_month = value.to_period('M')
    
    return year_month
```

```python
df['year_month'] = df['Date'].apply(get_year_month)
```

```python
df.groupby(['year_month', 'Company']).sum().groupby(level=0).cumsum()
```
