### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
data['Yr_Mo_Dy'].unique()
```

```python
datetime.date(2019, 3, 10)
```

```python
# 삼항연산자를 사용하여 연산 처리

def convert_date(x):
    year = x.year - 100 if x.year > 1989 else x.year
    date_log = datetime.date(year, x.month, x.day)
    
    return date_log

# convert datetime 
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(convert_date)
data.head()
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data['Yr_Mo_Dy'].dtype
```

```python
# conver columns type
# pd.to_datetime을 사용하면 datetime 형식으로 바꿀 수 있다.

data['Yr_Mo_Dy'] = pd.to_datetime(data['Yr_Mo_Dy'])
```

```python
# 이렇게도 바꿀 수 있다. 사실 이게 더 편하다.(알기만 하면...)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].astype('datetime64[ns]')
```
