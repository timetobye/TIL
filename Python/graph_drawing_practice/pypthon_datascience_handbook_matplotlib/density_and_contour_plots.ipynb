### 3차원 함수 시각화하기

A contour plot can be created with the ``plt.contour`` function.
It takes three arguments: a grid of *x* values, a grid of *y* values, and a grid of *z* values.
The *x* and *y* values represent positions on the plot, and the *z* values will be represented by the contour levels.
Perhaps the most straightforward way to prepare such data is to use the ``np.meshgrid`` function, which builds two-dimensional grids from one-dimensional arrays:

2차원 그리드 포인트 생성¶
변수가 2개인 2차원 함수의 그래프를 그리거나 표를 작성하려면 2차원 영역에 대한 (x,y) 좌표값 쌍 즉, 그리드 포인트(grid point)를 생성하여 각 좌표에 대한 함수 값을 계산해야 한다. 예를 들어 x, y 라는 두 변수를 가진 함수에서 x가 0부터 2까지, y가 0부터 4까지의 사각형 영역에서 변화하는 과정을 보고 싶다면 이 사각형 영역 안의 다음과 같은 그리드 포인트들에 대해 함수를 계산해야 한다.

(𝑥,𝑦)=(0,0),(0,1),(0,2),(0,3),(0,4),(1,0),⋯(2,4)
 
이러한 그리드 포인트를 만드는 과정을 도와주는 것이 meshgrid 명령이다. meshgrid 명령은 사각형 영역을 구성하는 가로축의 점들과 세로축의 점을 나타내는 두 벡터를 인수로 받아서 이 사각형 영역을 이루는 조합을 출력한다. 결과는 그리드 포인트의 x 값만을 표시하는 행렬과 y 값만을 표시하는 행렬 두 개로 분리하여 출력한다.

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
```

```python
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black')
```

데이터 범위 내에 20개의 동일한 간격으로 더 많은 선 그리기, 색상 지정

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar()
```

2차원 데이터 그리드를 이미지로 해석하는 plt.imshow() 함수를 사용

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

### imshow의 주의사항

- plt.imshow()는 x,y 그리드를 받지않으므로 플롯의 extent 를 직접 지정해야 한다.
- 기본적으로는 plt.imshow()는 대부분의 등고선 플롯과 같이 원점이 왼쪽 아래가 아니라 왼쪽 위에 위치한 표준 이미지 배열 정의를 따른다. 그리드 데이터를 보여줄 때는 이를 반드시 변경해야 함
- plt.imshow()는 입력 데이터에 매칭하기 위해 자동으로 축의 가로세로 비율을 조정한다.
  -  plt.axis(aspect='image')를 설정해 x와 y 단위를 일치시켜 바꿀 수 있다.

clabel을 이용하여 등고선 자체에 레이블을 추가할 거임 

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```
